```python
# 4(1)
import numpy as np

def sigmoid(x):
    return 1/(1 + np.exp(-x))

_input = np.array([[1],[2],[3]])
Winput_output = np.array([0.1,0.2,0.3])

output = np.dot(Winput_output,_input)
output = sigmoid(output)
print('4(1)')
print('출력노드의 출력 :', output)
# 4(2)
def target_output(target, output):
    return abs(target-output)

target = 1.0
err = target_output(target, output) # 0.19
e1 = err * 0.1 * (0.1+0.2+0.3)
e2 = err * 0.2 * (0.1+0.2+0.3)
e3 = err * 0.3 * (0.1+0.2+0.3)
print('4(2)')
print('x1으로 역전파되는 오차값 : ', e1)
print('x2으로 역전파되는 오차값 : ', e2)
print('x3으로 역전파되는 오차값 : ', e3)

# 4(3)
matrix_e = np.array([e1, e2, e3])
learning_rate = 0.1
wo_sum = np.dot(Winput_output,_input)
trans = -matrix_e*sigmoid(wo_sum)*(1-sigmoid(wo_sum))*output

trans = learning_rate*trans

print('4(3)')
print('Update된 가중치')
print('w1', Winput_output[0] - trans[0])
print('w2', Winput_output[1] - trans[1])
print('w3', Winput_output[2] - trans[2])
```
