```python
import numpy as np
from matplotlib import pyplot as plt

# h 기준값 정의 (1E-10)
h = 1E-10

# f1 정의
def f1(x):
    return x**2
# df1 정의
def df1(x):
    return (f1(x+h)-f1(x))/h
# f2 정의
def f2(x):
    return x**3 + x**2 + x
# df2 정의
def df2(x):
    return (f2(x+h)-f2(x))/h
# f3 정의
def f3(x):
    return x*np.sin(x*x*x+3)
# df3 정의
def df3(x):
    return (f3(x+h)-f3(x))/h
# f4 정의
def f4(x):
    return np.cos(np.sqrt(np.exp(x)+1)/2)
# df4 정의
def df4(x):
    return (f4(x+h)-f4(x))/h

# numpy 함수 활용하여 각 함수값 출력
print(' f1(2) =',round(f1(2),3))
print('df1(2) =',round(df1(2),3))
print(' f2(2) =',round(f2(2),3))
print('df2(2) =',round(df2(2),3))
print(' f3(2) =',round(f3(2),3))
print('df3(2) =',round(df3(2),3))
print(' f4(2) =',round(f4(2),3))
print('df4(2) =',round(df4(2),3))

# matplotlib 활용하여 도함수의 그래프 작성

_x = np.arange(-5,5,0.001)
plt.plot(_x,f1(_x),linestyle='solid', label='f1(x)')
plt.plot(_x,f2(_x),linestyle='solid', label='f2(x)')
plt.plot(_x,f3(_x),linestyle='solid', label='f3(x)')
plt.plot(_x,f4(_x),linestyle='solid', label='f4(x)')

plt.plot(_x,df1(_x),linestyle='solid', label='df1(x)')
plt.plot(_x,df2(_x),linestyle='solid', label='df2(x)')
plt.plot(_x,df3(_x),linestyle='solid', label='df3(x)')
plt.plot(_x,df4(_x),linestyle='solid', label='df4(x)')

plt.ylim(-30,30)
plt.legend()
```
