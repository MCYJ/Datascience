```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
irisdf = pandas.DataFrame(iris.data,columns=iris.feature_names)
irisdf['target']=iris.target
irisdf['target']=irisdf['target'].map({0:"setosa",1:"versicolor",2:"virginica"})

SL = irisdf.iloc[:,0]
SW = irisdf.iloc[:,1]
PL = irisdf.iloc[:,2]
PW = irisdf.iloc[:,3]
name = irisdf.iloc[:,4]


#1 학습 예제 외 붓꽃 데이터셋의 특성 중 가장 의미 있는 관계 가시화
# 꽃받침의 길이와 꽃잎의 넓이의 관계가 산포도로 표현했을때 서로 비례한다는 것을 알게 되었습니다.
plt.scatter(SL,PW, c=iris.target)
```

```python
#2 평균 가시화
# 각 특성별 그룹화를 하여 비교하는 것이 아이리스 종류별 특성차이를 파악하는데에 용이함
labels = ['setosa','versicolor','virginca']
_setosa = irisdf[irisdf['target']=='setosa'].mean(axis=0).values
_versicolor = irisdf[irisdf['target']=='versicolor'].mean(axis=0).values
_virginca = irisdf[irisdf['target']=='virginica'].mean(axis=0).values
index = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
width = 0.1
x = [0,1,2,3]
x2 = [0.1,1.1,2.1,3.1]
x3 = [0.2,1.2,2.2,3.2]

plt.bar(x, _setosa, width, label='setosa')
plt.bar(x2, _versicolor, width, label='versicolor')
plt.bar(x3, _virginca, width, label='virginica')
plt.xticks(x,index)
plt.legend(labels)
```

```python
#3 데이터셋 분할
# 아이리스 종류별로 먼저 분할 한 뒤, 각각의 데이터프레임에서 일부씩 분할하여 합친뒤 이를 csv파일로 변환
# 아이리스 종류별로 분할
test_setosa = irisdf[irisdf['target']=='setosa'][:13]
train_setosa = irisdf[irisdf['target']=='setosa'][13:]
test_versicolor = irisdf[irisdf['target']=='versicolor'][:13]
train_versicolor = irisdf[irisdf['target']=='versicolor'][13:]
test_virginca = irisdf[irisdf['target']=='virginica'][:13]
train_virginca = irisdf[irisdf['target']=='virginica'][13:]

# 종류별로 분할된 아이리스 결합
test = pd.concat([test_setosa, test_versicolor, test_virginca])
train = pd.concat([train_setosa, train_versicolor, train_virginca])

# test_dataset.csv train_dataset.csv 생성
test.to_csv('test_dataset.csv')
train.to_csv('train_dataset.csv')
```

```python
#4 분할된 데이터셋 가시화
# label, index 설정
labels = ['setosa','versicolor','virginca']
index = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']

# csv 파일로 부터 test/train 데이터 불러오기
test = pd.read_csv('test_dataset.csv')
train = pd.read_csv('train_dataset.csv')

test_setosa = test[test['target']=='setosa'].loc[:,index].mean(axis=0).values
test_versicolor = test[test['target']=='versicolor'].loc[:,index].mean(axis=0).values
test_virginca = test[test['target']=='virginica'].loc[:,index].mean(axis=0).values

train_setosa = train[train['target']=='setosa'].loc[:,index].mean(axis=0).values
train_versicolor = train[train['target']=='versicolor'].loc[:,index].mean(axis=0).values
train_virginca = train[train['target']=='virginica'].loc[:,index].mean(axis=0).values

# plot 생성
width = 0.1
x = [0,1,2,3]
x2 = [0.1,1.1,2.1,3.1]
x3 = [0.2,1.2,2.2,3.2]

plt.bar(x, test_setosa, width, label='setosa')
plt.bar(x2, test_versicolor, width, label='versicolor')
plt.bar(x3, test_virginca, width, label='virginica')
plt.xticks(x,index)
plt.legend(labels)
plt.figure()
plt.bar(x, train_setosa, width, label='setosa')
plt.bar(x2, train_versicolor, width, label='versicolor')
plt.bar(x3, train_virginca, width, label='virginica')
plt.xticks(x,index)
plt.legend(labels)

# 아래의 plot을 서로 비교하면 데이터가 비슷하다는 것을 알 수 있습니다.
```
