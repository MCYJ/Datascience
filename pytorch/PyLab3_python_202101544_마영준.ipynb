```python
import numpy as np
import matplotlib.pyplot as plt

wb = [3,76]

data = [[2,81],[4,93],[6,91],[8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

def predict(x):
    return wb[0]*x + wb[1]

def rmse(a, p):
    return np.sqrt(((a-p)**2).mean())

def rmse_val(predict_result,y):
    return rmse(np.array(predict_result), np.array(y))

predict_result = []

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))

print("rmse 최종값: "+ str(rmse_val(y, predict_result)))

print("기울기 a=", wb[0])
print("y 절편 b=", wb[1])
```

```python
def step_gradient(w_current, b_current, points, learning_rate):
    w_gradient = 0
    b_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i][0]
        y = points[i][1]
        w_gradient += -(2/N) * x * (y - (w_current * x + b_current))
        b_gradient += -(2/N) * (y - (w_current * x + b_current))
    new_w = w_current - (learning_rate * w_gradient)
    new_b = b_current - (learning_rate * b_gradient)
    return [new_w, new_b]
# 초기 w,b 설정
w = wb[0]
b = wb[1]
# step 1000 회
plt.title("Plotting the first 10 steps")
for i in range(1, 1001):
    w,b = step_gradient(w, b, data, 0.03)
    if i <= 10:
        plt.plot(learning_time, w*learning_time+b, label=i)
plt.legend()
learning_time = np.array([2,3,4,5,6,7,8])
#plt.plot(learning_time, w*learning_time+b)
print("w:",w,"b:",b)
```
